```python
## Loading the FIle
import fitz
doc = fitz.open("LenderFeesWorksheetNew.pdf")
```

```python
doc.metadata
```

```python
# Getting raw text from pdf
page = doc[0]
text = page.get_text("text")
print(text[:500])
```

```python
page = doc[0]
words = page.get_text("words")
for word in words[:5]:
    print(word)
```

```python
structured_list = []
for word in words:
  dic = {}
  dic['text'] = word[4]
  dic['bbox'] = list(word[0:4])
  structured_list.append(dic)
structured_list[-5:]
```

```python
import json

# Convert the structured_list to a JSON string
json_data = json.dumps(structured_list, indent=4)

# Define the filename for the JSON output
output_filename = "structured_list.json"

# Save the JSON data to a file
with open(output_filename, "w") as f:
    f.write(json_data)

print(f"Structured list successfully saved to {output_filename}")
```
